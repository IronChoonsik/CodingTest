```python
#DashInsert 함수
#숫자로 구성된 문자열을 입력받은 뒤, 문자열 안에서 홀수가 연속되면 두 수 사이에 -추가, 짝수가 연속되면 *를 추가하는 기능
#ex 4546793
#454*67-9-3

data = input("숫자를 입력해 주세요: ")
numbers = list(map(int,data))
result = []
for i, num in enumerate(numbers):
    result.append(str(num))
    
    if i  <len(numbers) -1: #다음 수가 있다면
        is_odd = num % 2 ==1
        is_next_odd = numbers[i+1] %2 ==1
        if is_odd and is_next_odd:
            result.append("-")
            
    elif not is_odd and not is_next_odd:
        result.append("*")
    
print(f'"{"".join(result)}"')
#print('"'+"".join(result)+'"')
```

```python
#DashInsert 함수
#숫자로 구성된 문자열을 입력받은 뒤, 문자열 안에서 홀수가 연속되면 두 수 사이에 -추가, 짝수가 연속되면 *를 추가하는 기능
#ex 4546793
#454*67-9-3

def DashInsert():
    data = input("숫자를 입력하세요 :") #12345 (str 상태임)
    numbers = list(map(int,data))#[1,2,3,4,5] len=5
    result=[]
    for i, num in enumerate(numbers): # enumerate 순서와 해당값이 같이 반환됨
        result.append(str(num)) #빈 리스트에 첫번째 숫자 넣어주기(i=0에 해당하는 num값) 이제 다음 수가 있는지 여부를 판단 ! 
                           #str 처리 안해주면expected str instance, int found 에러뜸
        if i < len(numbers) - 1: #i는 0,1,2,3,4 num은 1,2,3,4,5 만약 -1을 안해주면, 아래 if문에서 i가 4일 경우 i+1값을 확인 못해서 에러남
            is_odd = num %2 ==1 #num값 (지금 for문 돌려서 result에 추가한 숫자의 나머지가 1일경우)
            is_next_odd = numbers[i+1] %2 ==1 #(지금 for문 돌려서 result에 추가한 숫자의 다음 숫자의 나머지가 1일 경우)
            if is_odd and is_next_odd: #is_odd와 is_next_odd가 전부 참값이면 ( 둘다 홀수면)
                result.append("-") #result에다가 -를 넣어준다 
        elif not is_odd and not is_next_odd: #현재 숫자와 다음 숫자가 전부 나머지가 1이 아니면?(모두 짝수면?)
                result.append("*") #result에다가 *를 넣어준다
    print(f'"{"".join(result)}"')  # ''.join(리스트)는 ["a","b","c"]인 리스트를 'abc'로 반환해줌
    

a=DashInsert()
a   
```

```python
#0~9의 문자로 된 숫자를 입력받았을 때, 이 입력값이 0~9의 모든 숫자를 각각 한번씩만 사용한 것인지?

def dupnum(s):
    result = []
    for num in s:
        if num not in result:
            result.append(num)
        else:
            return False
        
    return len(result) == 10

dupnum("01234556789")
```
